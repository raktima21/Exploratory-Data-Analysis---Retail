### Analysis of columns 'Sales' and 'Profit'

#### With respect to column 'Ship Mode'

```python
x=df.groupby("Ship Mode").sum()
x1=x.drop(columns=["Quantity","Discount","Postal Code"])
x1.iloc[:,:3].plot(subplots=True,sharex=False,layout=(2,2),figsize=(12,12))
plt.show()
```

#### It can be clearly seen that both profit and sales are highest for Standard class and least for Same Day.

#### Analysis of the columns 'Sales','Profit' and 'Discount' with respect to column 'Category'

```python
plt.figure(figsize=(20,3))
state=df.groupby('Category').sum().reset_index()
state_sale=state.sort_values(by=("Sales"),ascending=False)
ed=sns.barplot(x="Sales",y="Category",data=state_sale)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('Sales',fontsize=20)
plt.ylabel('Category',fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(20,3))
state=df.groupby('Category').sum().reset_index()
state_sale=state.sort_values(by=("Profit"),ascending=False)
ed=sns.barplot(x="Profit",y="Category",data=state_sale)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('Profit',fontsize=20)
plt.ylabel('Category',fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(20,3))
state=df.groupby('Category').sum().reset_index()
state_sale=state.sort_values(by=("Discount"),ascending=False)
ed=sns.barplot(x="Discount",y="Category",data=state_sale)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('Discount',fontsize=20)
plt.ylabel('Category',fontsize=20)
plt.show()
```

#### It can be observed that both the amount of sales and profit are maximum for technology. However for category furniture , sale is quite high yet profit is least for it. The low profit is may be due to the high amount of discount provided for this category. 

#### Analysis of the columns 'Sales','Profit' and 'Discount' with respect to column 'Sub-Category'

```python
plt.figure(figsize=(20,10))
state=df.groupby('Sub-Category').sum().reset_index()
state_sale=state.sort_values(by=("Sales"),ascending=False)
ed=sns.barplot(x="Sales",y="Sub-Category",data=state_sale)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('Sales',fontsize=20)
plt.ylabel('Sub-Category',fontsize=20)
plt.show()
```

#### It can be clearly observed that Sale is highest for Phones and almost equally high for chairs. While Sale is very low for Labels and Fasteners.

```python
plt.figure(figsize=(20,10))
state=df.groupby('Sub-Category').sum().reset_index()
state_sale=state.sort_values(by=("Profit"),ascending=False)
ed=sns.barplot(x="Profit",y="Sub-Category",data=state_sale)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('Profit',fontsize=20)
plt.ylabel('Sub-Category',fontsize=20)
plt.show()
```

#### It can be clearly observed from the plot that the profit is highest for Copiers followed by Phones. While negative profits are observed for sub-cateories Supplies, Bookcases and Tables.

```python
plt.figure(figsize=(20,10))
state=df.groupby('Sub-Category').sum().reset_index()
state_sale=state.sort_values(by=("Discount"),ascending=False)
ed=sns.barplot(x="Discount",y="Sub-Category",data=state_sale)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('Discount',fontsize=20)
plt.ylabel('Sub-Category',fontsize=20)
plt.show()
```

#### A major reason for the negative profit observed for sub-categories tables and bookcases can be the high amount of discounts provided in them.

#### Analysis of the columns 'Sales','Profit' and 'Discount' with respect to column 'States'

```python
plt.figure(figsize=(30,18))
ed=sns.countplot(y="State",data=df)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('Count',fontsize=20)
plt.ylabel('State',fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(30,18))
state=df.groupby('State').sum().reset_index()
state_sale=state.sort_values(by=("Sales"),ascending=False)
ed=sns.barplot(x="Sales",y="State",data=state_sale)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('Sales',fontsize=20)
plt.ylabel('State',fontsize=20)
plt.show()
```

#### It can be seen that the sales are very high for states California, New York, Texas etc. While the sales are very low for states like Maine,West Virginia,North Dakota etc. 

```python
plt.figure(figsize=(30,18))
state=df.groupby('State').sum().reset_index()
state_sale=state.sort_values(by=("Quantity"),ascending=False)
ed=sns.barplot(x="Quantity",y="State",data=state_sale)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('Quantity',fontsize=20)
plt.ylabel('State',fontsize=20)
plt.show()
```

#### It can be seen that quantities are sold most in states like California, New York, Texas while the least are sold in states like Wyoming, West Virginia, North Dakota.

```python
plt.figure(figsize=(30,18))
state=df.groupby('State').sum().reset_index()
state_sale=state.sort_values(by=("Profit"),ascending=False)
ed=sns.barplot(x="Profit",y="State",data=state_sale)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('State',fontsize=20)
plt.ylabel('Profit',fontsize=20)
plt.show()
```

#### States like California, New York, Washington are giving very high profits. While states like Texas, Ohio, Pennsylvania, Illinois, North Carolina, Colorado, Tennessee, Arizona, Florida, Oregon, Wyoming and West Virginia are providing negative profit to the business. However sales are quite good in states like Texas, Ohio, Pennsylvania, Illionois and North Carolina.

```python
plt.figure(figsize=(30,18))
state=df.groupby('State').sum().reset_index()
state_sale=state.sort_values(by=("Discount"),ascending=False)
ed=sns.barplot(x="Discount",y="State",data=state_sale)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('State',fontsize=20)
plt.ylabel('Discount',fontsize=20)
plt.show()
```

#### States like Texas, Pennsylvania, Illionois, Ohio, California, Florida, North Carolina provide very high discount in their items.
